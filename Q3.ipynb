## Q3: How has BLM incidence impact on numbers of Burglary?

```python
import pandas as pd
import datetime
burglary_crime = pd.read_csv('Burglary_Crime_Q2.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plot_all = burglary_crime.groupby(['Month','Year']).size().reset_index(name='counts')
plot_all = plot_all[(plot_all['Year'] == 2017) | (plot_all['Year'] == 2018) | (plot_all['Year'] == 2019) | (plot_all['Year'] == 2020)]
sns.relplot(x='Month', y='counts', data=plot_all, hue="Year", kind="line")
plt.title('BLM & Burglary')
plt.xlabel('Month')
plt.ylabel('Number of Burglaries')
plt.vlines(5.9, 500, 1000, linestyles='dashed', label='BLM')
# Save the plot to a file
plt.savefig('BLM_Burglary.png', bbox_inches='tight')
```
