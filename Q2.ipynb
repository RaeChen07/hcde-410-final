## Q2: How has COVID impact on numbers of Burglary?

Use pandas and datetime to add column 'Year' and 'Month' to the dataframe and save it to a new csv fime 'Burglary_Crime_Q2.csv'.

```python
import pandas as pd
import datetime
burglary_crime = pd.read_csv('Burglary_Crime_Q1.csv')
burglary_crime['Year'] = burglary_crime['Offense Start DateTime'].apply(lambda x: x[:4])
burglary_crime['Month'] = burglary_crime['Offense Start DateTime'].apply(lambda x: x[5:7])
burglary_crime = burglary_crime[['Report Number','Offense Start DateTime','Offense Parent Group','Offense','day_of_week','Year','Month']]
burglary_crime.to_csv('Burglary_Crime_Q2.csv')
```

Use matplotlib and seaborn to plot the graph.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plot_all = burglary_crime.groupby(['Month','Year']).size().reset_index(name='counts')
plot_all = plot_all[(plot_all['Year'] == '2017') | (plot_all['Year'] == '2018') | (plot_all['Year'] == '2019') | (plot_all['Year'] == '2020')]
sns.relplot(x='Month', y='counts', data=plot_all, hue="Year", kind="line")
plt.title('COVID & Burglary')
plt.xlabel('Month')
plt.ylabel('Number of Burglaries')
plt.vlines(2.75, 500, 1000, linestyles='dashed', label='Stay at Home')
# Save the plot to a file
plt.savefig('COVID_Burglary.png', bbox_inches='tight')
```
